```python
import pandas as pd
import re

# Function to extract time in HH:MM format
def extract_time(val):
    if pd.isna(val):
        return None
    val = str(val).strip()
    
    # Match "8h 30m", "08h00m", "8 h 5 m", etc.
    match = re.match(r'(\d{1,2})\s*h\s*(\d{1,2})\s*m', val)
    if match:
        h, m = match.groups()
        return f"{int(h):02}:{int(m):02}"
    
    # Match "08:30", "8:5", etc.
    match = re.match(r'(\d{1,2}):(\d{1,2})', val)
    if match:
        h, m = match.groups()
        return f"{int(h):02}:{int(m):02}"
    
    return None

# Apply the function to initial_departure
new_dfbus['initial_departure'] = new_dfbus['initial_departure'].apply(extract_time)
```

```python
my_cursor.execute("DROP TABLE IF EXISTS busdetail")

my_cursor.execute("""
    CREATE TABLE IF NOT EXISTS busdetail (
        id INT AUTO_INCREMENT PRIMARY KEY,
        busname VARCHAR(255),
        route_name VARCHAR(255),
        bustype VARCHAR(255),
        departure_time VARCHAR(255),
        total_duration VARCHAR(255),
        reaching_time VARCHAR(255),
        star_rating FLOAT,
        price DECIMAL(10, 2),
        seats_available INT,
        route_link TEXT
    )
""")
mydb.commit()

sql_query = """
    INSERT INTO busdetail (
        busname, route_name, bustype, departure_time, total_duration, reaching_time,
        star_rating, price, seats_available, route_link
    ) VALUES (
        %s, %s, %s, %s, %s, %s, %s, %s, %s, %s
    );
"""
new_dfbus['Seats_Available'] = pd.to_numeric(new_dfbus['Seats_Available'], errors='coerce').astype('Int64')
new_dfbus['Seats_Available'] = new_dfbus['Seats_Available'].fillna(0).astype(int)
new_dfbus['Ratings'] = new_dfbus['Ratings'].fillna(0).astype(float)


new_dfbus['Price'] = new_dfbus['Price'].astype(str).str.replace('INR', '', regex=False).str.replace(',', '').str.strip()
new_dfbus['Price'] = pd.to_numeric(new_dfbus['Price'], errors='coerce')


for index, row in new_dfbus.iterrows():
    my_cursor.execute(sql_query, (
        None if pd.isna(row['Bus_name']) else row['Bus_name'],
        None if pd.isna(row['Route_name']) else row['Route_name'],
        None if pd.isna(row['Bus_type']) else row['Bus_type'],
        None if pd.isna(row['initial_departure']) else row['initial_departure'],
        None if pd.isna(row['Total_duration']) else row['Total_duration'],
        None if pd.isna(row['destination_arrival']) else row['destination_arrival'],
        None if pd.isna(row['Ratings']) else row['Ratings'],
        None if pd.isna(row['Price']) else row['Price'],
        None if pd.isna(row['Seats_Available']) else int(row['Seats_Available']),
        None if pd.isna(row['Route_link']) else row['Route_link']
    ))


mydb.commit()
```
